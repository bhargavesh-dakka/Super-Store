### Task 3: Find out the per unit price from the data

```python
total_price = df.Sales
number = df.Quantity
price_per_unit = total_price/number
df.insert(19,"Price per Unit",price_per_unit,True)
df.head()
```

### Task 4: Find out the monthly revenue and analyze the findings 

```python
df_2 = df
df_2['Month'] = df['Order Date'].dt.to_period('M')
df_2.sort_values(by='Month', inplace=True)
monthly_profit = df_2.groupby("Month")
unique_months = df_2.Month.unique()
df_2 = df_2.groupby(["Month"]).sum()
df_2.insert(3,"Month",unique_months,True)
df_2=df_2[['Month','Sales']]
df_2.head()
```

```python
df_2['Month'] = df_2['Month'].dt.strftime("%Y-%m")
plt.bar(df_2.Month,df_2.Sales)
```

#### Analysis:

Firstly, the choice of the bar graph was simple; we needed to see the values for each month and just compare them, ideal for a bar chart. It should be noted that the labels on the x-axis are not visible clearly since there are 48 different labels, and increasing the size of the graph for better visual representation is not something I have been able to explore yet. Nevertheless, it is very much possible to analyse the data.


Secondly, even though the graph that has been plotted is not big enough for more precise analysis, one can say that the monthly revenues were much higher near the end of the year compared to the start or even middle of the year. The significant spikes near the end of each year indicares as such.

### Task 5: Find out the yearly revenue and analyze the findings

```python
df_3 = df
df_3['Year'] = df['Order Date'].dt.to_period('Y')
df_3.sort_values(by='Year', inplace=True)
yearly_profit = df_3.groupby("Year")
unique_years = df_3.Year.unique()
df_3 = df_3.groupby(["Year"]).sum()
df_3.insert(3,"Year",unique_years,True)
df_3=df_3[['Year','Sales']]
df_3.head()
```

```python
df_3['Year'] = df_3['Year'].dt.strftime("%Y")
plt.bar(df_3.Year,df_3.Sales, color='green')
```

#### Analysis:

Firstly, same as the previous task, the choice of the bar graph was simple; we needed to see the values for each year and just compare them, ideal for a bar chart.

Secondly, there is a clear increase in the revenue in the later years, with the year 2017 seeing a huge rise. The year 2015 did see a slight dip, but the following years got the data back into the increasing trajectory that was expected with the natural rise in preferance, popularity, and loyalty.

### Task 6: Finding out the Monthly Growth Rate and analyse the findings

```python
length = len(df_2['Sales'])
mgr = [0.0]*length
for n in range(1,length):
    mgr[n] = ((df_2.Sales[n]-df_2.Sales[n-1])/df_2.Sales[n-1])*100
df_2.insert(2,"Monthly Growth Rate (%)",mgr,True)
df_2.head()
```

```python
max_mgr = df_2["Monthly Growth Rate (%)"].max()
print("Maximum Monthly Growth Rate(%) = " + str(max_mgr))
```

```python
plt.plot(df_2.Month,df_2["Monthly Growth Rate (%)"],color='orange')
```

#### Analysis:

The choice for this graph was a simple line plot since it would help analyse the difference between the rate of growth for monthly income, as such being ideal for such analysis. It should be noted that the labels on the x-axis are not visible clearly since there are 48 different labels, and increasing the size of the graph for better visual representation is not something I have been able to explore yet. Nevertheless, it is very much possible to analyse the data.

We can see clearly there is a massive peak in monthly income growth rate between the second and third month. Naturally, this peak is by far the highest since growth of the sales is most effective at the beginning of any venture, which is true for this particular data set as well.

### Task 7: Find out the most and least sold product ID

```python
df_4 = df
df_4.sort_values(by='Product ID', inplace=True)
unique_pid = df_4['Product ID'].unique()
df_4 = df_4.groupby(["Product ID"]).sum()
df_4.insert(3,"Product ID",unique_pid,True)
df_4=df_4[['Product ID','Quantity']]
df_4
```

```python
max_pid = df_4.loc[df_4['Quantity'].idxmax()]
print("Maxiumum Sold Product: \n" + str(max_pid))
min_pid = df_4.loc[df_4['Quantity'].idxmin()]
print("\nMinimum Sold Product: \n" + str(min_pid))
```

### Task 10: Finding out the majority and minority customer cities on basis of

```python
df_8 = df
df_8.sort_values(by='City', inplace=True)
unique_city = df_8['City'].unique()
df_9 = df_8.groupby(["City"]).count()
df_9.insert(0,"City",unique_city,True)
df_9=df_9[['City','Customer ID']]
df_9
```

```python
max_cyc = df_9.loc[df_9['Customer ID'].idxmax()]
print("City with most number of customers: \n" + str(max_cyc))
min_cyc = df_9.loc[df_9['Customer ID'].idxmin()]
print("\nCity with least number of customers: \n" + str(min_cyc))

# Here, one can clearly notice that the city with least number of customers is not just Aberdeen. The above tabel shows that
# cities like Abilene and Yucaipa also has the least number of customers. However, the idxmin() only outputs the first row with
# the minimum value, and as such, only Aberdeen is the output.
```

```python
df_8=df_8.groupby(["City"]).sum()
df_8.insert(0,"City",unique_city,True)
df_8=df_8[['City','Quantity','Sales']]
df_8
```

```python
max_cys = df_8.loc[df_8['Sales'].idxmax()]
print("City with highest sales value: \n" + str(max_cys))
min_cys = df_8.loc[df_8['Sales'].idxmin()]
print("\nCity with lowest sales value: \n" + str(min_cys))
```

```python
max_cyq = df_8.loc[df_8['Quantity'].idxmax()]
print("City with highest quantity sold: \n" + str(max_cyq))
min_cyq = df_8.loc[df_8['Quantity'].idxmin()]
print("\nCity with lowest quantity sold: \n" + str(min_cyq))
```
